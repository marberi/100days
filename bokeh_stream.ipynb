# Streaming video from the Tello drone

```python
import time
import cv2

from bokeh.plotting import figure
from bokeh.io import show, output_notebook, push_notebook

import djitellopy
```

```python
tello = djitellopy.Tello()
tello.connect()

tello.streamon()
frame_read = tello.get_frame_read()
```

```python
frame_in = frame_read.frame
frame = cv2.cvtColor(frame_in, cv2.COLOR_BGR2RGBA)
frame = cv2.flip(frame, -1)

width, height, _ = frame_in.shape

output_notebook()
p = figure(x_range=(0,width), y_range=(0,height), output_backend='webgl')
myImage = p.image_rgba([frame], 0, 0, width, height)

show(p, notebook_handle=True)

while True:
    print('pushing...')
    frame = cv2.cvtColor(frame_read.frame, cv2.COLOR_BGR2RGBA)
    frame = cv2.flip(frame, -1)
    myImage.data_source.data['image'] = [frame]
    push_notebook()
    
    time.sleep(0.1)
```
